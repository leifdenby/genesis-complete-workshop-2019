# Introduction

In this notebook you'll be using [xarray](http://xarray.pydata.org/en/stable/) to open and explore the simulation output from a Large-Eddy simulation of shallow convection. The case is based around the [RICO](http://projects.knmi.nl/rico/) observational campaign which studied precipitating shallow convection over the ocean in the tropical Atlantic. In difference to the published case the surface fluxes have been fixed and the initial ambient horizontal wind is set to zero (to remove the effect of shear on cloud development). The simulation was done over a $20km \times 20km \times 4km$ with isotropic resolution $\Delta x = 25m$ with the [UCLALES](https://github.com/uclales/uclales) model.

![Image credits: Bjorn Stevens](images/rico_rainbow_stevens.gif)

The files provided contain the instantaneous 3D fields 6hrs into the simulation (once the atmosphere has reach a quasi steady-state) and have been clipped to only extend up to 1.5km height to reduce the file sizes (the original simulation extended to 4km).

*NOTE*: If you are unfamiliar with `xarray` you should read through the [quick overview](http://xarray.pydata.org/en/stable/quick-overview.html). By working with *named* coordinates (instead of simply accessing array dimensions through indecies `0,1,...`) `xarray` provides a much more intuitive interface to loading N-dimensional array datasets and it [supports many different file formats](http://xarray.pydata.org/en/stable/installing.html#for-netcdf-and-io).

```python
# load the modules we'll need

import xarray as xr
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# open all 3D files into one xarray Dataset
ds = xr.open_mfdataset('../fixed_flux_noshear__nx800//3d_blocks/full_domain/*.nc', combine='by_coords')
ds
```

```python
# have a look at one variable
ds.w
```

```python
# extract a cross-section of vertical velocity mid boundary-layer (z~400m) and plot
# NOTE: velocities are stored on cell-edges (xm, ym, zm) instead of cell-centers
ds.sel(zm=400, method='nearest').w.plot()
```

```python
# extract a vertical cross-section of vertical velocity
fig, ax = plt.subplots(figsize=(16, 2))
ax.set_aspect(1)
ds.sel(xt=200, method='nearest').w.plot(y='zm', ax=ax)
```

```python
# extract a vertical cross-section so we can see clouds
fig, ax = plt.subplots(figsize=(16, 2))
ax.set_aspect(1)
ds.sel(xt=200, method='nearest').l.plot(y='zt', ax=ax)
```

```python
# check mean profile of cloud liquid so we can see where the clouds are
ds.mean(dim=('xt', 'yt'), keep_attrs=True).l.plot(y='zt')
```
