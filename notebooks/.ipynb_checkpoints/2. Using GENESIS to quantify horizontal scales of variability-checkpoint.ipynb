In this notebook you will use the GENESIS toolkit to extract length-scales of variable of different scalar fields

```python
# load the modules we'll need

import xarray as xr
import matplotlib.pyplot as plt

import genesis
import genesis.length_scales
import genesis.utils

%matplotlib inline

genesis.__version__
```

```python
# open all 3D files into one xarray Dataset
ds = xr.open_mfdataset('../fixed_flux_noshear__nx800/3d_blocks/full_domain/*.nc', combine='by_coords')
```

```python
# center all the staggered velocity fields (otherwise we won't be able 
# to look at their covariance with other fields)
ds['u'] = genesis.utils.center_staggered_field(ds.u)
ds['v'] = genesis.utils.center_staggered_field(ds.v)
ds['w'] = genesis.utils.center_staggered_field(ds.w)
```

## Covariance in a single cross-section

```python
# look at the covariance of water vapour at z~300m
ds_slice = ds.sel(zt=300, method='nearest').squeeze()
v1 = ds_slice.q
v2 = ds_slice.q

C_qq = genesis.length_scales.cumulant.calc.calc_2nd_cumulant(v1=v1, v2=v2)
C_qq.plot()
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(12, 4))

# extract length-scales
ds_slice = ds.sel(zt=300, method='nearest').rename(dict(xt='x', yt='y')).squeeze()
v1 = ds_slice.q
v2 = ds_slice.q

genesis.length_scales.cumulant.calc.covariance_plot(ax=axes[0], v1=v1, v2=v2, log_scale=False)
genesis.length_scales.cumulant.calc.covariance_direction_plot(ax=axes[1], v1=v1, v2=v2)
plt.tight_layout()
```

```python
ds_scales = genesis.length_scales.cumulant.calc.charactistic_scales(v1=v1, v2=v2)
ds_scales
```

```python
ds_scales.width_principle
```

## Profiles of length-scales

```python
# extract coherence length-scale for water vapour with height

v1_3d = ds.q
v2_3d = ds.q

ds_lengths_qq = genesis.length_scales.cumulant.vertical_profile.calc.get_height_variation_of_characteristic_scales(
    v1_3d=v1_3d, v2_3d=v2_3d, z_max=600., 
)
```

```python
ds_lengths_qq
```

```python
ds_lengths_qq.width_perpendicular.plot(y='zt', label='$L_\perp$')
ds_lengths_qq.width_principle.plot(y='zt', label='$L_0$')
plt.legend()
plt.xlim(0, None)
plt.xlabel('width [m]')
```

```python
# extract coherence length-scale for vertical velocity with height

v1_3d = ds.w
v2_3d = ds.w

ds_lenghts_ww = genesis.length_scales.cumulant.vertical_profile.calc.get_height_variation_of_characteristic_scales(
    v1_3d=v1_3d, v2_3d=v2_3d, z_max=600., 
)
```

```python
ds_lenghts_ww.width_perpendicular.plot(y='zt', label='$L_\perp$')
ds_lenghts_ww.width_principle.plot(y='zt', label='$L_0$')
plt.legend()
plt.xlim(0, None)
plt.xlabel('width [m]')
```

```python
# Comparing the vertical and water vapour scales
ds_lengths_qq.width_principle.plot(y='zt', label='$L_0(q)$')
ds_lenghts_ww.width_principle.plot(y='zt', label='$L_0(w)$')
plt.legend()
plt.xlim(0, None)
plt.xlabel('width [m]')
```

Notice how the vertical velocity features are much narrower than those in water vapour - is this what you expected based on the previous exercise?
